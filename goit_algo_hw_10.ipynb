Вам необхідно написати дві функції для касової системи, яка видає решту покупцеві:

Функція жадібного алгоритму find_coins_greedy. Ця функція повинна приймати суму, яку потрібно видати покупцеві, і повертати словник із кількістю монет кожного номіналу, що використовуються для формування цієї суми. Наприклад, для суми 113 це буде словник {50: 2, 10: 1, 2: 1, 1: 1}. Алгоритм повинен бути жадібним, тобто спочатку вибирати найбільш доступні номінали монет.
Функція динамічного програмування find_min_coins. Ця функція також повинна приймати суму для видачі решти, але використовувати метод динамічного програмування, щоб знайти мінімальну кількість монет, необхідних для формування цієї суми. Функція повинна повертати словник із номіналами монет та їх кількістю для досягнення заданої суми найефективнішим способом. Наприклад, для суми 113 це буде словник {1: 1, 2: 1, 10: 1, 50: 2}
Порівняйте ефективність жадібного алгоритму та алгоритму динамічного програмування, базуючись на часі їх виконання або О великому та звертаючи увагу на їхню продуктивність при великих сумах. Висвітліть, як вони справляються з великими сумами та чому один алгоритм може бути більш ефективним за інший у певних ситуаціях. Свої висновки додайте у файл readme.md домашнього завдання.

Greedy Algorithm

````python
def find_coins_greedy(amount):
    coins = [50, 25, 10, 5, 2, 1]
    result = {}

    for coin in coins:
        if amount >= coin:
            count = amount // coin
            amount -= count * coin
            result[coin] = count

    return result

# Example
amount = 113
print(find_coins_greedy(amount))  # Output: {50: 2, 10: 1, 2: 1, 1: 1}
````

Dynamic programming algorithm

````python
def find_min_coins(amount):
    coins = [1, 2, 5, 10, 25, 50]
    dp = [float('inf')] * (amount + 1)
    dp[0] = 0
    coin_used = [0] * (amount + 1)

    for i in range(1, amount + 1):
        for coin in coins:
            if i >= coin and dp[i - coin] + 1 < dp[i]:
                dp[i] = dp[i - coin] + 1
                coin_used[i] = coin

    result = {}
    while amount > 0:
        coin = coin_used[amount]
        if coin in result:
            result[coin] += 1
        else:
            result[coin] = 1
        amount -= coin

    return result

# Example
amount = 113
print(find_min_coins(amount))  # Output: {1: 1, 2: 1, 10: 1, 50: 2}
````

Порівняння ефективності
Жадібний алгоритм:
Часова складність: O(n), де n - кількість номіналів монет.
Просторова складність: O(1).
Переваги: Швидкий та простий у реалізації, підходить для великих сум, коли номінали монет дозволяють швидко наблизитись до суми.
Недоліки: Може не знайти оптимальне рішення, якщо набір монет містить номінали, які не дозволяють легко досягнути заданої суми.
Алгоритм динамічного програмування:
Часова складність: O(n * m), де n - кількість номіналів монет, m - сума, яку потрібно видати.
Просторова складність: O(m).
Переваги: Гарантує знаходження оптимального рішення з мінімальною кількістю монет.
Недоліки: Може бути повільнішим для великих сум, особливо якщо діапазон можливих сум великий.

Висновки
Обидва алгоритми мають свої переваги та недоліки, які визначають їх ефективність у різних ситуаціях:

Жадібний алгоритм підходить для систем, де швидкість виконання критична, і номінали монет дозволяють швидко наближатися до потрібної суми. Він особливо корисний у реальних касових апаратах, де потрібно швидко видавати решту.
Алгоритм динамічного програмування гарантує мінімальну кількість монет для видачі решти, але може бути повільнішим. Він підходить для систем, де точність важливіша за швидкість, і суми, з якими потрібно працювати, не дуже великі.

````python

Це домашнє завдання спрямоване на розробку та порівняння двох алгоритмів для видачі решти: жадібного алгоритму та алгоритму динамічного програмування.

1. **Функція жадібного алгоритму**: `find_coins_greedy`
   - Приймає суму, яку потрібно видати покупцеві.
   - Повертає словник із кількістю монет кожного номіналу.
   - Використовує жадібний підхід для вибору найбільших доступних номіналів монет.

2. **Функція динамічного програмування**: `find_min_coins`
   - Приймає суму для видачі решти.
   - Використовує метод динамічного програмування для знаходження мінімальної кількості монет.
   - Повертає словник із номіналами монет та їх кількістю для досягнення заданої суми найефективнішим способом.

- **Часова складність:** O(n)
- **Просторова складність:** O(1)
- **Переваги:** Швидкий та простий у реалізації.
- **Недоліки:** Може не знайти оптимальне рішення для всіх наборів монет.

- **Часова складність:** O(n * m)
- **Просторова складність:** O(m)
- **Переваги:** Гарантує оптимальне рішення.
- **Недоліки:** Може бути повільнішим для великих сум.

- **Жадібний алгоритм** ефективний для великих сум, коли потрібно швидко видати решту.
- **Алгоритм динамічного програмування** забезпечує мінімальну кількість монет, але може бути менш ефективним за часом для великих сум.

Для запуску функцій використовуйте наведені приклади:

```python
# Greedy Algorythm
amount = 113
print(find_coins_greedy(amount))  # Output: {50: 2, 10: 1, 2: 1, 1: 1}

# Dynamic programming Algorythm
print(find_min_coins(amount))  # Output: {1: 1, 2: 1, 10: 1, 50: 2}
````
